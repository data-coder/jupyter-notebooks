# Triangular Surface Plot

A triangular surface plot is formed by placing surfaces in place of the triangles formed by points in a 3D space. All triangles form a mesh. 

The trisurf method is used to plot a 3D surface with a triangular mesh.

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D # <--- This is important for 3d plotting 
```

```python
X = [1, 2, 1]
Y = [0, 0.5, 1]
Z = [0, 0.5, 1]
```

Any three points, when non-collinear, determine a unique triangle and simultaneously, a unique plane.

```python
fig = plt.figure()
ax = fig.gca(projection="3d")
ax.scatter3D(X, Y, Z)
```

```python
fig = plt.figure()
ax = fig.gca(projection="3d")
ax.plot_trisurf(X, Y, Z)
```

```python
X = [1, 2, 1]
Y = [0, 0.5, 1]
Z = [0, 0, 0]
fig = plt.figure()
ax = fig.gca(projection="3d")
ax.plot_trisurf(X, Y, Z)
```

Let's add another point 2
