```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

## 2D Model

```python
price = np.arange(100, 200)
sales_predicted = np.array([1000 - 2*p for p in price])
sales_observed = np.array([1000 - 2*p + r 
                           for p,r in zip(price, np.random.normal(0, 10, 100))])
```

```python
fig, ax = plt.subplots(figsize=(5, 5))
sns.scatterplot(x=price, y=sales_observed)
ax.set_xlabel('price')
ax.set_ylabel('sales')
```

```python
fig, ax = plt.subplots(figsize=(5, 5))
sns.scatterplot(x=price, y=sales_observed)
sns.lineplot(x=price, y=sales_predicted, color='red')
ax.set_xlabel('price')
ax.set_ylabel('sales')
```

## 3D Model

```python
price = np.arange(100, 200)
advertising_expenditure = np.random.normal(1000, 10, 100)
sales_predicted = np.array([1000 - 2*p + 0.6*a
               for p, a in zip(price, advertising_expenditure)])
sales_observed = np.array([1000 - 2*p + 0.6*a + r
                 for p, a, r in zip(price,
                                    advertising_expenditure,
                                    np.random.normal(0, 20, 100))])
```

```python
fig, ax = plt.subplots(figsize=(5, 5))
sns.scatterplot(x=price, y=sales_observed)
ax.set_xlabel('price')
ax.set_ylabel('sales')
plt.savefig('/tmp/statistical-learning-sales-price.png')
```

```python
fig, ax = plt.subplots(figsize=(5, 5))
sns.scatterplot(x=price, y=sales_observed)
sns.lineplot(x=price, y=sales_predicted, color='red')
ax.set_xlabel('price')
ax.set_ylabel('sales')
plt.savefig('/tmp/statistical-learning-sales-price.png')
```

```python
fig, ax = plt.subplots(figsize=(5, 5))
sns.scatterplot(x=advertising_expenditure, y=sales_observed)
ax.set_xlabel('advertising expenditure')
ax.set_ylabel('sales')
plt.savefig('/tmp/statistical-learning-sales-advertising.png')
```

```python
fig, ax = plt.subplots(figsize=(5, 5))
sns.scatterplot(x=advertising_expenditure, y=sales_observed)
sns.lineplot(x=advertising_expenditure, y=sales_predicted, color='red')
ax.set_xlabel('advertising expenditure')
ax.set_ylabel('sales')
plt.savefig('/tmp/statistical-learning-sales-advertising.png')
```

```python
fig, ax = plt.subplots(figsize=(10, 10))
ax = plt.axes(projection="3d")
ax.scatter3D(price, advertising_expenditure, sales_observed)
ax.set_xlabel('price')
ax.set_ylabel('advertising expenditure')
ax.set_zlabel('sales')
plt.savefig('/tmp/statistical-learning-sales-price-advertising.png')
```
