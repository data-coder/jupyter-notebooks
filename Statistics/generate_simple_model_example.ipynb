## 2D Model

Model with only one variable

```python
price = np.arange(100, 200)
sales_predicted = np.array([1000 - 2 * p for p in price])
sales_observed = np.array(
    [1000 - 2 * p + r for p, r in zip(price, np.random.normal(0, 10, 100))]
)
```

### Plots

```python
plot1 = ('scatterplot', price, sales_observed, fplots.FBLUE)
fplots.multiple_plots('Price', 'Sales', '', plot1)
```

```python
fig, ax = plt.subplots(figsize=(5, 5))
sns.scatterplot(x=price, y=sales_observed)
sns.lineplot(x=price, y=sales_predicted, color="red")
ax.set_xlabel("price")
ax.set_ylabel("sales")
```

```python
# the same using fplot module
TEMPPLOT1 = ('scatterplot', price, sales_observed, fblue)
TEMPPLOT2 = ('lineplot', price, sales_predicted, fred)
fplots.multiple_plots('Price', 'Sales', '', *[TEMPPLOT1, TEMPPLOT2])
```

## 3D Model

```python
price = np.arange(100, 200)
advertising_expenditure = np.random.normal(1000, 10, 100)
sales_predicted = np.array(
    [1000 - 2 * p + 0.6 * a for p, a in zip(price, advertising_expenditure)]
)
sales_observed = np.array(
    [
        1000 - 2 * p + 0.6 * a + r
        for p, a, r in zip(price, advertising_expenditure, np.random.normal(0, 20, 100))
    ]
)
```

```python
fig, ax = plt.subplots(figsize=(5, 5))
sns.scatterplot(x=price, y=sales_observed)
ax.set_xlabel("price")
ax.set_ylabel("sales")
plt.savefig("/tmp/statistical-learning-sales-price.png")
```

```python
fig, ax = plt.subplots(figsize=(5, 5))
sns.scatterplot(x=price, y=sales_observed)
sns.lineplot(x=price, y=sales_predicted, color="red")
ax.set_xlabel("price")
ax.set_ylabel("sales")
plt.savefig("/tmp/statistical-learning-sales-price.png")
```
