## 2D Model

Model with only one variable

```python
price = np.arange(100, 200)
sales_predicted = np.array([1000 - 2 * p for p in price])
sales_observed = np.array(
    [1000 - 2 * p + r for p, r in zip(price, np.random.normal(0, 10, 100))]
)
```

### Plots

```python
plot1 = ('scatterplot', price, sales_observed, fplots.FBLUE)
fplots.multiple_plots('Price', 'Sales', '', plot1, logo=LOGO)
```

```python
fig, ax = plt.subplots(figsize=(5, 5))
sns.scatterplot(x=price, y=sales_observed)
sns.lineplot(x=price, y=sales_predicted, color="red")
ax.set_xlabel("price")
ax.set_ylabel("sales")
```

```python
fig, ax = plt.subplots(figsize=(5, 5))
sns.scatterplot(x=price, y=sales_observed)
sns.lineplot(x=price, y=sales_predicted, color="red")
ax.set_xlabel("price")
ax.set_ylabel("sales")
ax.text(ax.get_xlim()[0], ax.get_ylim()[0], 'datacomy')
```

## 3D Model

```python
price_2 = np.arange(100, 200)
advertising_expenditure_2 = np.random.normal(1000, 10, 100)
sales_predicted_2 = np.array(
    [1000 - 2 * p + 0.6 * a for p, a in zip(price, advertising_expenditure_2)]
)
sales_observed_2 = np.array(
    [
        1000 - 2 * p + 0.6 * a + r
        for p, a, r in zip(price, advertising_expenditure_2, np.random.normal(0, 20, 100))
    ]
)
```

```python
fig, ax = plt.subplots(figsize=(5, 5))
sns.scatterplot(x=price_2, y=sales_observed_2)
ax.set_xlabel("price")
ax.set_ylabel("sales")
plt.savefig("/tmp/statistical-learning-sales-price.png")
```

```python
fig, ax = plt.subplots(figsize=(5, 5))
sns.scatterplot(x=price, y=sales_observed_2)
sns.lineplot(x=price_2, y=sales_predicted_2, color="red")
ax.set_xlabel("price")
ax.set_ylabel("sales")
plt.savefig("/tmp/statistical-learning-sales-price.png")
```

```python
fig, ax = plt.subplots(figsize=(5, 5))
sns.scatterplot(x=advertising_expenditure_2, y=sales_observed_2)
ax.set_xlabel("advertising expenditure")
ax.set_ylabel("sales")
plt.savefig("/tmp/statistical-learning-sales-advertising.png")
```

```python
fig, ax = plt.subplots(figsize=(5, 5))
sns.scatterplot(x=advertising_expenditure_2, y=sales_observed_2)
sns.lineplot(x=advertising_expenditure_2, y=sales_predicted_2, color="red")
ax.set_xlabel("advertising expenditure")
ax.set_ylabel("sales")
plt.savefig("/tmp/statistical-learning-sales-advertising.png")
```
