The purpose of meshgrid is to create a rectangular grid out of an array of x values and an array of y values.

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D  # <--- This is important for 3d plotting 
```

```python
A = [1, 2, 3]
B = [4, 5, 6, 7]
```

```python
J = np.meshgrid(A, B)
J
```

For surface plots, a meshgrid is needed.

Now, I'll add the z values. Since we usually will have observations in the form (X, Y, Z), the vectors to be meshed are of the same size. This means that the meshgrid will be a square and not a rectangle, like in the above example.

In this example, I'll use 1x3 vectors. Let's assume we have only three observations:
Observation 1: (Feature1=1, Feature=4, value=10)
Observation 2: (Feature1=2, Feature=5, value=11)
Observation 3: (Feature1=3, Feature=6, value=12)

Then, the vectors are:

```python
D = [1, 2, 3]
E = [4, 5, 6]
Z_values = [10, 11, 12]
```

```python
K = np.meshgrid(D, E)
K
```

Firs, we need to create a grid on the "base" of the plot:

```python
fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(D, [0, 0, 0], color='b', s=100)
ax.scatter([0, 0, 0], E, color='r', s=100)
ax.scatter(K[0], K[1], color='g', s=100)
ax.set_xlim([0, 4])
ax.set_ylim([0, 8])
ax.set_zlim([0, 15])
```

The grid has nine points: 3 features times 3 features.

Now I add the Z values. We need nine points. If all values were 10, we'd had:

```python
fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(D, [0, 0, 0], color='b', s=100)
ax.scatter([0, 0, 0], E, color='r', s=100)
ax.scatter(K[0], K[1], color='g', s=100)
ax.scatter(K[0], K[1], [10, 10, 10, 10, 10, 10, 10, 10, 10], color='black', s=100)
ax.set_xlim([0, 4])
ax.set_ylim([0, 8])
ax.set_zlim([0, 15])
```

```python
fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(D, [0, 0, 0], color='b', s=100)
ax.scatter([0, 0, 0], E, color='r', s=100)
ax.scatter(K[0], K[1], color='g', s=100)
ax.scatter(K[0], K[1], [10, 10, 10, 10, 10, 10, 10, 10, 10], color='black', s=10)
ax.plot(D, E, Z_values, color='black')
ax.set_xlim([0, 4])
ax.set_ylim([0, 8])
ax.set_zlim([0, 15])
```
